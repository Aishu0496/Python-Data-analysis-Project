```python
# Preprocess the data
Data_df['DATE'] = pd.to_datetime(Data_df['DATE'])
Data_df['MONTH'] = Data_df['DATE'].dt.to_period('M')

# Calculate total sales revenue per store per month
total_sales_revenue = Data_df.groupby(['STORE_NBR', 'MONTH'])['TOT_SALES'].sum().reset_index()

# Calculate total number of customers per store per month
total_customers = Data_df.groupby(['STORE_NBR', 'MONTH'])['LYLTY_CARD_NBR'].nunique().reset_index()
total_customers.rename(columns={'LYLTY_CARD_NBR': 'TOTAL_CUSTOMERS'}, index=str, inplace=True)

# Calculate average number of transactions per customer per store per month
transactions_per_customer = Data_df.groupby(['STORE_NBR', 'MONTH'])['TXN_ID'].nunique().reset_index()
transactions_per_customer.rename(columns={'TXN_ID': 'TOTAL_TRANSACTIONS'}, index=str, inplace=True)
transactions_per_customer = pd.merge(transactions_per_customer, total_customers, on=['STORE_NBR', 'MONTH'])
transactions_per_customer['AVG_TXNS_PER_CUSTOMER'] = transactions_per_customer['TOTAL_TRANSACTIONS'] / transactions_per_customer['TOTAL_CUSTOMERS']

# Merge all metrics into a single DataFrame
store_monthly_metrics = pd.merge(total_sales_revenue, total_customers, on=['STORE_NBR', 'MONTH'])
store_monthly_metrics = pd.merge(store_monthly_metrics, transactions_per_customer[['STORE_NBR', 'MONTH', 'AVG_TXNS_PER_CUSTOMER']], on=['STORE_NBR', 'MONTH'])

store_monthly_metrics.head()
```

```python
from scipy.stats import pearsonr

def compare_stores(trial_store, control_stores, metrics_df, metric_columns):
    results = []
    trial_store_metrics = metrics_df[metrics_df['STORE_NBR'] == trial_store]

    for control_store in control_stores:
        control_store_metrics = metrics_df[metrics_df['STORE_NBR'] == control_store]
        merged_metrics = pd.merge(trial_store_metrics, control_store_metrics, on='MONTH', suffixes=('_trial', '_control'))

        pearson_corrs = {}
        for metric in metric_columns:
            # Check if both trial and control have enough data for the current metric
            if len(merged_metrics[f'{metric}_trial']) >= 2 and len(merged_metrics[f'{metric}_control']) >= 2:
                pearson_corr, _ = pearsonr(merged_metrics[f'{metric}_trial'], merged_metrics[f'{metric}_control'])
                pearson_corrs[metric] = pearson_corr

            else:
                pearson_corrs[metric] = None

        magnitude_distances = {}
        for metric in metric_columns:
            observed_distance = abs(merged_metrics[f'{metric}_trial'] - merged_metrics[f'{metric}_control']).mean()
            min_distance = abs(merged_metrics[f'{metric}_trial'] - merged_metrics[f'{metric}_trial']).mean()
            max_distance = abs(merged_metrics[f'{metric}_trial'] - metrics_df[metric].mean()).max()
            magnitude_distance = 1 - (observed_distance - min_distance) / (max_distance - min_distance)
            magnitude_distances[metric] = magnitude_distance

        results.append({
            'trial_store': trial_store,
            'control_store': control_store,
            'pearson_corrs': pearson_corrs,
            'magnitude_distances': magnitude_distances
        })

    return results

# Define trial stores and control stores
trial_stores = [77, 86, 88]
control_stores = store_monthly_metrics['STORE_NBR'].unique()
control_stores = [store for store in control_stores if store not in trial_stores]

# Metrics to compare
metric_columns = ['TOT_SALES', 'TOTAL_CUSTOMERS', 'AVG_TXNS_PER_CUSTOMER']

# Compare control stores to each trial store
comparison_results = {}
for trial_store in trial_stores:
    comparison_results[trial_store] = compare_stores(trial_store, control_stores, store_monthly_metrics, metric_columns)

# Display comparison results for one of the trial stores
comparison_results[77][:5]  # Displaying first 5 results for trial store 77 for brevity
```

####Best Control Store

```python
best_control_stores = {}

for trial_store, results in comparison_results.items():
    best_store = None
    best_score = -1
    for result in results:
        score = sum(result['magnitude_distances'].values())
        if score > best_score:
            best_score = score
            best_store = result['control_store']
    best_control_stores[trial_store] = best_store

best_control_stores
```

###Compare matrics during the trial period

Assume the trial period is from 2019-01 to 2019-03.



```python
trial_period = pd.period_range('2019-01', '2019-03', freq='M')

trial_vs_control = []

for trial_store, control_store in best_control_stores.items():
    trial_data = store_monthly_metrics[(store_monthly_metrics['STORE_NBR'] == trial_store) & (store_monthly_metrics['MONTH'].isin(trial_period))]
    control_data = store_monthly_metrics[(store_monthly_metrics['STORE_NBR'] == control_store) & (store_monthly_metrics['MONTH'].isin(trial_period))]

    trial_vs_control.append({
        'trial_store': trial_store,
        'control_store': control_store,
        'trial_data': trial_data,
        'control_data': control_data
    })

trial_vs_control[0]  # Displaying the first comparison for brevity
```

###Statistical testing

```python
from scipy.stats import ttest_ind

stat_test_results = []

for comparison in trial_vs_control:
    trial_data = comparison['trial_data']
    control_data = comparison['control_data']

    ttest_results = {}
    for metric in metric_columns:
        t_stat, p_val = ttest_ind(trial_data[metric], control_data[metric])
        ttest_results[metric] = {'t_stat': t_stat, 'p_val': p_val}

    stat_test_results.append({
        'trial_store': comparison['trial_store'],
        'control_store': comparison['control_store'],
        'ttest_results': ttest_results
    })

stat_test_results[0]  # Displaying the first statistical test result for brevity
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plotting environment
sns.set(style="whitegrid")

# Function to plot metrics for a trial and control store
def plot_store_metrics(trial_store, control_store, metrics_df, trial_period):
    metrics = ['TOT_SALES', 'TOTAL_CUSTOMERS', 'AVG_TXNS_PER_CUSTOMER']
    trial_data = metrics_df[metrics_df['STORE_NBR'] == trial_store]
    control_data = metrics_df[metrics_df['STORE_NBR'] == control_store]

    # Convert 'MONTH' column to datetime if it's not already
    # Use .to_timestamp() to handle PeriodDtype
    trial_data['MONTH'] = trial_data['MONTH'].dt.to_timestamp()
    control_data['MONTH'] = control_data['MONTH'].dt.to_timestamp()

    fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(12, 18))

    for i, metric in enumerate(metrics):
        ax = axes[i]
        sns.lineplot(x='MONTH', y=metric, data=trial_data, ax=ax, label='Trial Store')
        sns.lineplot(x='MONTH', y=metric, data=control_data, ax=ax, label='Control Store')
        ax.axvline(pd.Timestamp('2019-01-01'), color='red', linestyle='--')
        ax.axvline(pd.Timestamp('2019-03-31'), color='red', linestyle='--')
        ax.set_title(f'Monthly {metric.replace("_", " ").title()}')
        ax.set_xlabel('Month')
        ax.set_ylabel(metric.replace("_", " ").title())
        ax.legend()

    plt.tight_layout()
    plt.show()

# Plot metrics for a trial and its best control store
plot_store_metrics(77, best_control_stores[77], store_monthly_metrics, trial_period)
```

###Compare Metrics During the Trial Period

```python
def plot_trial_period_comparison(trial_vs_control, metric_columns):
    for comparison in trial_vs_control:
        trial_store = comparison['trial_store']
        control_store = comparison['control_store']
        trial_data = comparison['trial_data']
        control_data = comparison['control_data']

        fig, axes = plt.subplots(nrows=1, ncols=len(metric_columns), figsize=(18, 6))

        for i, metric in enumerate(metric_columns):
            ax = axes[i]
            sns.barplot(x=['Trial Store', 'Control Store'],
                        y=[trial_data[metric].mean(), control_data[metric].mean()],
                        ax=ax)
            ax.set_title(f'Average {metric.replace("_", " ").title()}')
            ax.set_ylabel(metric.replace("_", " ").title())

        plt.suptitle(f'Comparison During Trial Period (Store {trial_store} vs Store {control_store})')
        plt.tight_layout(rect=[0, 0, 1, 0.96])
        plt.show()

# Plot comparison during the trial period
plot_trial_period_comparison(trial_vs_control, metric_columns)
```

###Visualize Statistical Test Results

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_stat_test_results(stat_test_results, metric_columns):
    for result in stat_test_results:
        trial_store = result['trial_store']
        control_store = result['control_store']
        ttest_results = result['ttest_results']

        fig, axes = plt.subplots(nrows=1, ncols=len(metric_columns), figsize=(18, 6))

        for i, metric in enumerate(metric_columns):
            ax = axes[i]
            trial_mean = store_monthly_metrics[(store_monthly_metrics['STORE_NBR'] == trial_store) & (store_monthly_metrics['MONTH'].isin(trial_period))][metric].mean()
            control_mean = store_monthly_metrics[(store_monthly_metrics['STORE_NBR'] == control_store) & (store_monthly_metrics['MONTH'].isin(trial_period))][metric].mean()
            p_val = ttest_results[metric]['p_val']

            sns.barplot(x=['Trial Store', 'Control Store'],
                        y=[trial_mean, control_mean],
                        ax=ax,
                        palette=['blue', 'green'])  # Specify your desired colors here

            ax.set_title(f'{metric.replace("_", " ").title()} (p-value: {p_val:.3f})')
            ax.set_ylabel(metric.replace("_", " ").title())

        plt.suptitle(f'Statistical Test Results (Store {trial_store} vs Store {control_store})')
        plt.tight_layout(rect=[0, 0, 1, 0.96])
        plt.show()

# Plot statistical test results
plot_stat_test_results(stat_test_results, metric_columns)
```
